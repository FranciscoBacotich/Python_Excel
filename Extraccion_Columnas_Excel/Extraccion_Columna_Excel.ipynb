Ahora vamos a crear un template con un versionado de nuestro archivo.
En el loop, creamos un nuevo df en base a la columna de nuestro data frame original que coincida con el state que extrae el loop. Asi iteramos sobre la columna y copiamos el campo en base a que coincida con el df original. Evitamos dedups y copiar campos fantasmas

Cuando creamos el archivo excel apartir de nuestro nuevo df, podemos ahi especificar una tab especifica. Lo mismo al importar.Tambien sumamos index = false para sacarle el ID

```python
# blank template: codigo para crear nuevo archivo
writer = pd.ExcelWriter("Splited Raw Data.xlsx", engine = 'xlsxwriter')

# Llevar registro de nombres usados para evitar duplicados
used_gender = set()

# Usando la logica del loop para extraer columna
for state in df['gender'].unique():
    # Limpiar caracteres no válidos y cortar si es muy largo
    clean_gender = re.sub(r'[\\/*?:\[\]]', "", str(state))[:31]

     # Evitar duplicados agregando un número
    count = 1
    original_gender = clean_gender
    while clean_gender in used_gender:
        clean_gender = f"{original_gender[:28]}_{count}"
        count += 1

    used_gender.add(clean_gender)

    # Escribir hoja
    newDF = df[df['gender'] == state]
    newDF.to_excel(writer, sheet_name = clean_gender, index = False)

writer.close()
```

El resultado final es una hoja de excel con tabs/hojas individuales por en este caso genero.
